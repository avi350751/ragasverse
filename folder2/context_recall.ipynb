### LLM Based Context Recall

```python
from ragas import SingleTurnSample
from ragas.metrics import LLMContextRecall
from ragas.llms import LangchainLLMWrapper
from langchain_openai import ChatOpenAI

test_case = SingleTurnSample(
    user_input="Who won the FIFA World Cup 2022?",
    response="Argentina was the champion.",
    reference="Argentina became champion defeating France in the finals.",
    retrieved_contexts=["Argentina became 3 times world cup champion in 2022.",
    "2022 World cup champion defeated France on penalties."],
)

evaluator_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4o-mini"))
context_recall = LLMContextRecall(llm=evaluator_llm)
await context_recall.single_turn_ascore(test_case)
```
